# How are in-demand skills trending for Data Analysts

```python
# importing libraries
import ast 
import pandas as pd 
from datasets import load_dataset 
import matplotlib.pyplot as plt 
import seaborn as sns

# loading the dataset
dataset = load_dataset('lukebarousse/data_jobs') 
df = dataset['train'].to_pandas() 

# data cleanup 
df['job_posted_date'] = pd.to_datetime(df['job_posted_date']) 
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_India = df[(df['job_title'] == 'Data Analyst') & (df['job_country'] == 'India')].copy()
```

```python
df_DA_India['job_posted_month_no'] = df_DA_India['job_posted_date'].dt.month
```

```python
df_DA_India_explode = df_DA_India.explode('job_skills')
```

```python
df_DA_India_pivot = df_DA_India_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

df_DA_India_pivot
```

```python
df_DA_India_pivot = df_DA_India_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

df_DA_India_pivot.loc['Total'] = df_DA_India_pivot.sum()

df_DA_India_pivot = df_DA_India_pivot[df_DA_India_pivot.loc['Total'].sort_values(ascending=False).index]

df_DA_India_pivot = df_DA_India_pivot.drop('Total')
```

```python
df_DA_India_pivot
```

```python
DA_Totals = df_DA_India.groupby('job_posted_month_no').size()

DA_Totals
```

```python
df_DA_India_perc = df_DA_India_pivot.div(DA_Totals/100, axis=0)

df_DA_India_perc
```

```python
df_DA_India_perc = df_DA_India_perc.reset_index()
df_DA_India_perc['job_posted_month'] = (df_DA_India_perc['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b')))
df_DA_India_perc = df_DA_India_perc.set_index('job_posted_month')
df_DA_India_perc = df_DA_India_perc.drop(columns='job_posted_month_no')

df_DA_India_perc
```

```python
df_plot = df_DA_India_perc.iloc[:, :5]

sns.set_theme(style='ticks')
sns.lineplot(data=df_plot, dashes=False, palette='tab10')
sns.despine()

plt.title('Trending Top Skills for Data Analysts in India')
plt.ylabel('Likelihood in Job postings')
plt.xlabel('Months')
plt.legend().remove()

from matplotlib.ticker import PercentFormatter
ax=plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))

for i in range(5):
  plt.text(11.2, df_plot.iloc[-1, i], df_plot.columns[i])
```
